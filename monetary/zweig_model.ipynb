# The trading algorithm

```python
overview = []
class TradingDecision:
    def __init__(self, date, close, decision_type, portfolio_value):
        self.date = date
        self.close = close
        self.decision_type = decision_type
        self.portfolio_value = portfolio_value
    
    def to_dict(self):
        return {
            "date": self.date,
            "S&P close": round(self.close, 2),
            "decision type": self.decision_type,
            "portfolio value": round(self.portfolio_value)
        }
```

```python
# Trading algo
funds = 10000
holdings = 0
prior_index = None

# meta
changes = []

def buy(index, row):
    global funds
    global holdings
    if holdings == 0:
        print("bought on " + str(row.name)[0:11] + " at a price of " + str(row["close"]))
        holdings = funds / row["close"]
        funds = 0
        print("funds are now " + str(funds))
        print("holdings are now " + str(holdings))
        print(" ")
        overview.append(TradingDecision(index, row["close"], "buy", holdings * row["close"]))

def sell(index, row):
    global funds
    global holdings
    if funds == 0:
        print("sold on " + str(row.name)[0:11] + " at a price of " + str(row["close"]))
        funds = row["close"] * holdings
        holdings = 0
        print("funds are now " + str(funds))
        print("holdings are now " + str(holdings))
        print(" ")
        overview.append(TradingDecision(index, row["close"], "sell", funds))
    
for index, row in trading_days.iterrows():
    if prior_index == None:
        buy(index, row)
        print(type(row))
    else:
        current_rate = row["rate"]
        old_rate = trading_days.loc[prior_index]["rate"]
        if current_rate != old_rate:
            # A change in the prime rate has occured
            change = current_rate - old_rate
            print("OLD RATE: " + str(old_rate))
            print("CURRENT RATE: " + str(current_rate))
            print("CHANGE: " + str(change))
            changes.append(change)
            if change < 0:
                # prime rate was cut -> BUY SIGNALS
                if old_rate < 5:
                    # buy signal 1 (any initial cut if peak was under 5%)
                    buy(index, row)
                else:
                    # buy signal 2 (either second cut or full 1 percentage point cut)
                    if change <= -1 or (len(changes) > 1 and changes[-2] < 0):
                        buy(index, row)
            else:
                # prime rate was hiked -> SELL SIGNALS
                if old_rate > 5:
                    # sell signal 1 (any initial hike if primes low was over 5%)
                    sell(index, row)
                else:
                    #sell signal 2 (second hike or full 1 percentage point hike)
                    if change >= 1 or (len(changes) > 1 and changes[-2] > 0):
                        sell(index, row)
                        
    prior_index = index

sell(trading_days.index[-1], trading_days.iloc[-1])

print("\nCash holdings at the end of the testing period:")
print(str(round(funds)) + "$")
```
