```python
print('These dates are in prime_cleansed, but not in spy:\n')
counter = 0
in_prime_cleansed_but_not_in_spy = list()

for index in prime_cleansed.index:
    if index not in spy.index:
        print(index)
        in_prime_cleansed_but_not_in_spy.append(index)
        counter += 1
        
print('\nTotal counts:', counter )
print('Difference of rows:', prime_cleansed.shape[0] - spy.shape[0])
```

```python
trading_days = spy.join(prime_cleansed, how="inner")
trading_days
```

```python
# Change rate type (str) to float
trading_days = trading_days.astype({"rate": 'float64'})
type(trading_days.rate[0])
```

```python
change_history = []
number_of_changes = 0
prior_index = None

for index, row in trading_days.iterrows():
    if prior_index != None:
        old_rate = trading_days.loc[prior_index]["rate"]
#         print(old_rate)
        if row["rate"] != old_rate:
            number_of_changes += 1
            change_history.append(row["rate"] - old_rate)
    prior_index = index

print(change_history)

np_change_history = np.array(change_history)
unique, counts = np.unique(np_change_history, return_counts=True)

print(unique, counts)
print(number_of_changes)
```

```python
plt.bar(unique, counts, width = 0.25) 
plt.xlabel("percentage point changes")
plt.ylabel("number of occurence")
plt.title("Prime Rate changes 1996 to 2021 (days with no changes removed)")

plt.show()
```
